```python
# Import library yang diperlukan
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score, KFold
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')
```

## Load Dataset

```python
print("Loading dataset...")
df = pd.read_csv('Iris.csv')
print(f"Ukuran dataset: {df.shape}")
```

### Dataset yang didapatkan ada 150 data dengan 6 dimensi

```python
print("\nInformasi Dataset:")
print(df.info())
```

### Setelah diketahui bahwa 6 dimensi tersebut merupakan Id, SepalLengthCm, SepalWidthCm, PetalLengthCm, PetalWidthCm, Species

```python
# Statistik deskriptif
print("\nStatistik Deskriptif:")
print(df.describe())
```

```python
# Daftar fitur
features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']

# Hitung mean dan std per kelas
summary = {}

for species in df['Species'].unique():
    species_data = df[df['Species'] == species]
    stats = {}
    for feature in features:
        mean = species_data[feature].mean()
        std = species_data[feature].std()
        stats[feature] = (round(mean, 2), round(std, 2))
    summary[species] = stats

# Tampilkan hasil dalam bentuk DataFrame
result_df = pd.DataFrame({
    species: {
        feature: f"{mu}, {sigma}"
        for feature, (mu, sigma) in stats.items()
    }
    for species, stats in summary.items()
}).T

# Ubah nama kolom agar lebih rapi
result_df.columns = [f"{col} (μ, σ)" for col in features]
result_df.index.name = "Kelas"

# Tampilkan tabel
print(result_df)
```

Data Preparation

```python
print("Mempersiapkan data...")
X = df.drop(['Id', 'Species'], axis=1)
y = df['Species']
```

Langkah Pertama dalam Machine learning adalah memisahkan mana label dan data latih, data yang kita miliki sudah jelas apabila Species merupakan label, dan data latih ya 4 dimensi penting yakni, `SepalLengthCm`, `SepalWidthCm`, `PetalLengthCm`, `PetalWidthCm`

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
print(f"Ukuran data training: {X_train.shape}")
print(f"Ukuran data testing: {X_test.shape}")
```

Data Latih terbagi menjadi 70:30 sehingga dari 150 data, data latih mendapatkan porsi sebanyak 105 data, dan sisanya yakni 45 data menjadi data uji.
Data dipisah secara random 

```python
# Scaling fitur (opsional untuk Naive Bayes, tetapi sering membantu)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Membangun model Naive Bayes
print("\nMembangun model Naive Bayes...")
nb_model = GaussianNB()

# Melatih model
print("Melatih model...")
nb_model.fit(X_train_scaled, y_train)
```

Disini kita menggunakan StandardScaler, agar mendapatkan perhitungan 0.0-1.0 dan memudahkan model dalam medapatkan value untuk dilatih

```python
# Evaluasi model dengan cross-validation
print("\nMengevaluasi model dengan cross-validation...")
cv = KFold(n_splits=5, shuffle=True, random_state=42)
cv_scores = cross_val_score(nb_model, X_train_scaled, y_train, cv=cv, scoring='accuracy')
print(f"Akurasi cross-validation (5-fold): {cv_scores}")
print(f"Rata-rata akurasi cross-validation: {cv_scores.mean():.4f} ± {cv_scores.std():.4f}")
```

Setelah dilakukan cross validation yakni merubah rubah data untuk dilatih, maka didapatkan rata rata akurasi sebesar 98%, hal ini mudah didapatkan karena data yang digunakan sebesar 150, apabila data semakin besar maka akurasinya juga akan turun diantara 70%-80%

```python
# Prediksi pada data testing
print("\nMelakukan prediksi pada data testing...")
y_pred = nb_model.predict(X_test_scaled)
```

```python

# Evaluasi performa model
print("\nHasil evaluasi model:")
print(f"Akurasi: {accuracy_score(y_test, y_pred):.4f}")

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

dari hasil ujicoba faktual kita mendapatkan nilai akurasi sebesar 0.91 atau 91% untuk memperediksi macam jenis bunga iris secara akurat

```python
print("\nConfusion Matrix:")
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=nb_model.classes_, 
            yticklabels=nb_model.classes_)
plt.xlabel('Prediksi')
plt.ylabel('Aktual')
plt.title('Confusion Matrix')
plt.savefig('confusion_matrix.png')
plt.show()
```

bisa kita lihat dari confusion matrix di atas bahwa setosa memiliki akurasi yang ketat, diikuti versicolor dan virginica. Hal ini bisa terlihat pada visualisasi data dimana data setosa jarang beririsan dengan versicolor dan virginica sehingga mesin pun otomatis dapat memisahkan setosa dengan mudah

```python
print("\nMenghitung probabilitas posterior...")
probs = nb_model.predict_proba(X_test_scaled)
prob_df = pd.DataFrame(probs, columns=nb_model.classes_)
print("\nProbabilitas posterior untuk 5 sampel pertama:")
print(prob_df.head())
```

```python
# Contoh implementasi: Fungsi prediksi untuk data baru
def predict_species(sepal_length, sepal_width, petal_length, petal_width):
    # Data baru
    new_data = np.array([[sepal_length, sepal_width, petal_length, petal_width]])
    
    # Scaling
    new_data_scaled = scaler.transform(new_data)
    
    # Prediksi
    prediction = nb_model.predict(new_data_scaled)[0]
    
    # Probabilitas
    probabilities = nb_model.predict_proba(new_data_scaled)[0]
    prob_dict = {class_name: prob for class_name, prob in zip(nb_model.classes_, probabilities)}
    
    return {
        'prediction': prediction,
        'probabilities': prob_dict
    }

# Contoh penggunaan
print("Contoh penggunaan fungsi prediksi:")
sample_1 = predict_species(5.1, 3.5, 1.4, 0.2)  # Kemungkinan Iris-setosa
print(f"Prediksi: {sample_1['prediction']}")
print(f"Probabilitas: {sample_1['probabilities']}")

sample_2 = predict_species(6.7, 3.0, 5.2, 2.3)  # Kemungkinan Iris-virginica
print(f"Prediksi: {sample_2['prediction']}")
print(f"Probabilitas: {sample_2['probabilities']}")
```

```python
# Simpan model untuk penggunaan masa depan
print("\nMenyimpan model ke disk...")
import pickle
with open('naive_bayes_iris_model.pkl', 'wb') as file:
    pickle.dump({
        'model': nb_model,
        'scaler': scaler,
        'classes': nb_model.classes_
    }, file)

print("\nModel berhasil disimpan sebagai 'naive_bayes_iris_model.pkl'")
print("\nProses RAD selesai!")
```

```python
# Memuat model
def load_model(model_path='naive_bayes_iris_model.pkl'):
    with open(model_path, 'rb') as file:
        model_data = pickle.load(file)
    return model_data

# Membuat prediksi
def predict_with_loaded_model(model_data, sepal_length, sepal_width, petal_length, petal_width):
    # Data baru
    new_data = np.array([[sepal_length, sepal_width, petal_length, petal_width]])
    
    # Scaling
    new_data_scaled = model_data['scaler'].transform(new_data)
    
    # Prediksi
    prediction = model_data['model'].predict(new_data_scaled)[0]
    
    # Probabilitas
    probabilities = model_data['model'].predict_proba(new_data_scaled)[0]
    prob_dict = {class_name: prob for class_name, prob in zip(model_data['classes'], probabilities)}
    
    return {
        'prediction': prediction,
        'probabilities': prob_dict
    }

# Contoh penggunaan
model_data = load_model('naive_bayes_iris_model.pkl')
result = predict_with_loaded_model(model_data, 5.1, 3.5, 1.4, 0.2)
print(f"Prediksi: {result['prediction']}")
print(f"Probabilitas: {result['probabilities']}")
# """)
```
